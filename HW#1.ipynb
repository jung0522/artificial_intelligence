#  privacy 보호방법

#Assignment conditions

```python
import cv2

image_path = '/content/drive/MyDrive/정준영.jpg'

def mosaic(image_path, x, y, w, h):

    image = cv2.imread(image_path)
    focusing = image[y:y+h, x:x+w]

    mosaic_focusing = cv2.resize(focusing, (10, 5), interpolation=cv2.INTER_NEAREST)
    mosaic_focusing = cv2.resize(mosaic_focusing, (w, h), interpolation=cv2.INTER_NEAREST)

    image[y:y+h, x:x+w] = mosaic_focusing

    cv2_imshow(image)

x, y, w, h = 350, 560, 600, 150

mosaic(image_path, x, y, w, h)
```
